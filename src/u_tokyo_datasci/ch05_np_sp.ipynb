# scipy

```python
x = np.linspace(0, 10, 11)
y = np.cos(-(x**2) / 5.0)
plt.scatter(x, y)
plt.grid(True)
```

```python
f = sp.interpolate.interp1d(x, y, "linear")
plt.plot(x, y, "o", x, f(x), "-")
plt.grid(True)
```

```python
# スプライン3次補完
f2 = sp.interpolate.interp1d(x, y, "cubic")
x2 = np.linspace(0, 10, 30)

plt.plot(x, y, "o", x, f(x), "-", x2, f2(x2), "--")
plt.legend(["data", "linear", "cubic"], loc="best")
plt.grid(True)
```

```python
# p139
x = np.linspace(0, 10, 11)
y = np.sin(x**2 / 5.0)

f = sp.interpolate.interp1d(x, y, "linear")
f2 = sp.interpolate.interp1d(x, y, "quadratic")
f3 = sp.interpolate.interp1d(x, y, "cubic")
x_inter = np.linspace(0, 10, 31)

plt.plot(
    x,
    y,
    "o",
    x_inter,
    f(x_inter),
    "-",
    x_inter,
    f2(x_inter),
    "--",
    x_inter,
    f3(x_inter),
    "-.",
)
plt.legend(["data", "linear", "quadratic", "cubic"], loc="best")
plt.grid(True)
```

```python
# 最適化
def f(x):
    y = 2 * x**2 + 2 * x - 10
    return y


x = np.linspace(-4, 4)
plt.plot(x, f(x))
plt.plot(x, np.zeros(len(x)))
plt.grid(True)
```

```python
# x=2付近
x = sp.optimize.fsolve(f, 2)
print(x)
```

```python
# min
def objective(x):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    x4 = x[3]
    return x1 * x4 * (x1 + x2 + x3) + x3


# s.t.
def constraint1(x):
    return x[0] * x[1] * x[2] * x[3] - 25.0


def constraint2(x):
    return 40.0 - x[0] ** 2 * x[1] ** 2 * x[2] ** 2 * x[3] ** 2


# 初期値
x0 = [1, 5, 5, 1]
# print(objective(x0))

# 制約の準備
b = (1.0, 5.0)
bnds = (b, b, b, b)
con1 = {"type": "ineq", "fun": constraint1}
con2 = {"type": "ineq", "fun": constraint2}
cons = [con1, con2]

# 実行
sol = sp.optimize.minimize(objective, x0, method="SLSQP", bounds=bnds, constraints=cons)
print(sol)
```

```python
print(f"when: {sol.x=}, min: {sol.fun=}")
```

```python
# p149
def f(x):
    return 5 * x - 10


x = np.linspace(-10, 10, 21)
plt.plot(x, f(x))
plt.plot(x, np.zeros(len(x)))
plt.grid(True)

sp.optimize.fsolve(f, 2.5)
```

```python
def f2(x):
    return x**3 - 2 * x**2 - 11 * x + 12


x = np.linspace(-10, 10, 21)
plt.plot(x, f2(x))
plt.plot(x, np.zeros(len(x)))
plt.grid(True)

print(sp.optimize.fsolve(f2, -3))
print(sp.optimize.fsolve(f2, 0))
print(sp.optimize.fsolve(f2, 5))
```

# 5章 総合問題

```python
# ∫[0, 1]∫[0, 1-x] f(x, y) dydx
sp.integrate.dblquad(
    lambda x, y: 1 / (np.sqrt(x + y) * (1 + x + y) ** 2), 0, 1, 0, lambda x: 1 - x
)
```

```python
def obj(x):
    return x**2 + 1


def con1(x):
    return x + 1


cons = [{"type": "ineq", "fun": con1}]
sol = sp.optimize.minimize(obj, 0, method="SLSQP", constraints=cons)
sol
```

```python
print(f"when: {sol.x=}, min: {sol.fun=}")
```
