- 結合

  - Inner Join: data1 と data2 の両方に存在するレコードだけ取り出して結合
  - Outer Join: data1 と data2 のレコードを全て結合
  - Left Join: data1 のレコードを全て取り出して、対応する data2 のレコードのみ結合
  - Right Join: data2 のレコードを全て取り出して、対応する data1 のレコードのみ結合

  - (縦結合: concat による縦の積み上げ)

## 欠損値

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.rand(10, 4))
df.iloc[1, 0] = np.nan
df.iloc[2, 2] = np.nan
df.iloc[5:, 3] = np.nan

df
```

```python
# リストワイズ削除: NaNを含む行を全て削除
df.dropna()
```

```python
# ペアワイズ削除: まず大きく欠損している列を削除して、利用可能なデータのみを使う
df[[0, 1]].dropna()
```

```python
df.fillna(0)
```

```python
# 直前の行の値で埋める
df.fillna(method="ffill")
```

```python
df.mean()
```

```python
# 平均値代入法
df.fillna(df.mean())
```

```python
?df.fillna
```

```python
fx_jpus.resample('ME').last()  # resampling: ある頻度のデータを別の頻度のデータで取り出し直す処理
```

```python
fx_jpus.resample('D').last()
```

```python
fx_jpus.resample('D').ffill().head()
```

```python
fx_jpus.head()
```

```python
fx_jpus.shift(1).head()
```

```python
shift1_ratio = fx_jpus / fx_jpus.shift(1)
shift1_ratio.head()
```

```python
fx_jpus.diff()  # A - B
```

```python
fx_jpus.pct_change()  # (A - B) / B
```
